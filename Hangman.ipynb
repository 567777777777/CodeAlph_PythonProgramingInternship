```python
import random

# Define hangman stages
hangman_stages = [
    """
    +---+
    |   |
        |
        |
        |
        |
    =========
    """,
    """
    +---+
    |   |
    O   |
        |
        |
        |
    =========
    """,
    """
    +---+
    |   |
    O   |
    |   |
        |
        |
    =========
    """,
    """
    +---+
    |   |
    O   |
   /|   |
        |
        |
    =========
    """,
    """
    +---+
    |   |
    O   |
   /|\\  |
        |
        |
    =========
    """,
    """
    +---+
    |   |
    O   |
   /|\\  |
   /    |
        |
    =========
    """,
    """
    +---+
    |   |
    O   |
   /|\\  |
   / \\  |
        |
    =========
    """
]

# List of words for the game
word_list = [
    "analyst",
    "engineer",
    "administrator",
    "machineengineer",
    "scientist",
    "architect",
    "statistician",
    "analyst",
    "manager"
]

# Select a random word from the word list
chosen_word = random.choice(word_list)

# Initialize the display to show underscores for each letter in the chosen word
display = ['_' if char.isalpha() else char for char in chosen_word]

# Initialize the number of lives
lives = len(hangman_stages) - 1

# Main game loop
while True:
    # Display the current hangman stage
    print(hangman_stages[len(hangman_stages) - 1 - lives])
    
    # Display the current state of the word
    print(' '.join(display))

    # Check if the player has guessed all the letters
    if '_' not in display:
        print("Congratulations! You guessed the word correctly!")
        break

    # Check if the player has run out of lives
    if lives == 0:
        print(f"Sorry, you're out of lives. The word was '{chosen_word}'. Better luck next time!")
        break

    # Get a letter guess from the player
    guessed_letter = input("Guess a letter: ").lower()

    # Check if the guessed letter is in the word
    if guessed_letter in chosen_word:
        for i in range(len(chosen_word)):
            if chosen_word[i] == guessed_letter:
                display[i] = guessed_letter
        print("Correct guess!")

    else:
        lives -= 1
        print("Incorrect guess. Try again!")

# End of game
print("Game Over.")
```
